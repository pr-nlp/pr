```python
def remove_stopwords(text):

  #lower text
  text = text.lower()

  #split text
  text = text.split()

  return text

remove_stopwords(sentence1)
```

```python
def remove_stopwords(text):

  #lower text
  text = text.lower()

  #split text
  text = text.split()

  for word in text:
    if word in stop_words:
      text.remove(word)

  return text

remove_stopwords(sentence3)
```
