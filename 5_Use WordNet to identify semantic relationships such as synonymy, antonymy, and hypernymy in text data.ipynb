```python
# WordNet Example in NLP
# Install:
# pip install nltk

import nltk
from nltk.corpus import wordnet

# Download WordNet data (run once)
nltk.download('wordnet')

# Input word
word = "boy"

# Get synsets
synsets = wordnet.synsets(word)

# -----------------------------
# SYNONYMS
# -----------------------------
print("Synonyms:")
synonyms = set()

for syn in synsets:
    for lemma in syn.lemmas():
        synonyms.add(lemma.name())

print(synonyms)

# -----------------------------
# ANTONYMS
# -----------------------------
print("\nAntonyms:")
antonyms = set()

for syn in synsets:
    for lemma in syn.lemmas():
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

print(antonyms)

# -----------------------------
# HYPERNYMS
# -----------------------------
print("\nHypernyms:")
hypernyms = set()

for syn in synsets:
    for hyper in syn.hypernyms():
        hypernyms.add(hyper.lemma_names()[0])

print(hypernyms)
```
