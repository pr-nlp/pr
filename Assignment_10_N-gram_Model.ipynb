Author Detection using N-Gram Model

```python
from collections import defaultdict, Counter
import re
```

```python
class NGramAuthorDetector:
    def __init__(self, n=2):
        self.n = n
        self.author_profiles = {}

    def preprocess(self, text):
        text = text.lower()
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        return text.split()

    def train(self, author, text):
        words = self.preprocess(text)

        ngrams = []

        for i in range(len(words) - self.n + 1):
            gram = tuple(words[i:i+self.n])
            ngrams.append(gram)

        self.author_profiles[author] = Counter(ngrams)

    def predict(self, text):
        words = self.preprocess(text)

        test_ngrams = []

        for i in range(len(words) - self.n + 1):
            gram = tuple(words[i:i+self.n])
            test_ngrams.append(gram)

        scores = {}

        for author, profile in self.author_profiles.items():
            score = 0

            for gram in test_ngrams:
                score += profile.get(gram, 0)

            scores[author] = score

        predicted_author = max(scores, key=scores.get)

        return predicted_author, scores
```

```python
# Training data for authors

author1_text = """
Machine learning is transforming the world.
Artificial intelligence is the future.
Deep learning provides powerful solutions.
"""

author2_text = """
Nature is beautiful and full of mystery.
Trees and rivers create peaceful environments.
Wildlife makes forests alive.
"""
```

```python
# Create model

model = NGramAuthorDetector(n=2)   # bigram model

# Train model

model.train("Tech Author", author1_text)
model.train("Nature Author", author2_text)
```

```python
# Test text

test_text = """
Artificial intelligence and deep learning are growing rapidly.
"""

prediction, scores = model.predict(test_text)

print("Predicted Author:", prediction)
print("Scores:", scores)
```

Speech Recognition Style N-Gram Model

```python
from collections import defaultdict, Counter
```

```python
class NGramSpeechRecognizer:
    def __init__(self, n=2):
        self.n = n
        self.ngrams = defaultdict(Counter)

    def train(self, sentences):
        for sentence in sentences:

            words = sentence.lower().split()

            for i in range(len(words) - self.n + 1):

                context = tuple(words[i:i+self.n-1])
                next_word = words[i+self.n-1]

                self.ngrams[context][next_word] += 1

    def predict_next_word(self, text, top_k=3):

        words = text.lower().split()

        context = tuple(words[-(self.n-1):])

        if context in self.ngrams:

            predictions = self.ngrams[context].most_common(top_k)

            return [word for word, count in predictions]

        return ["No prediction"]
```

```python
# Speech training data

speech_data = [
    "hello how are you",
    "hello how is everything",
    "how are you doing",
    "how is your day",
    "are you fine"
]
```

```python
# Train model

speech_model = NGramSpeechRecognizer(n=2)

speech_model.train(speech_data)
```

```python
# Predict next word

print(speech_model.predict_next_word("hello"))
print(speech_model.predict_next_word("how"))
```
