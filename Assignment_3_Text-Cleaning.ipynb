# Text cleaning:
**1. HTML Tags Cleaning**

```python
sample = '<p>Browsers usually insert quotation marks around the q element.</p><p>WWFs goal is to: <q>Build a future where people live in harmony with nature.</q></p>'
```

```python
import re
def striphtml(data):
  p = re.compile(r'<.*?>')
  return p.sub('', data)

```

```python
striphtml(sample)
```

**Encode the emoji into utf-8 format and decode the same to ignore ascii characters**

```python
import re
def clean_text(text):


    #lowering the text
    text = text.lower()

    #removing anything starts from @ #
    text = re.sub("[@|#]\w+","",text)

    #encoding emoji's
    text = text.encode('utf-8')
    text = text.decode('utf-8', 'ignore')


     #removing unicode text format
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    return text

clean_text(tweet)
```

**Removal of URLs**

```python
import re
def remove_url(text):
  pattern = re.compile(r'https?://\S+|www\.\S+')
  return pattern.sub(r'', text)
```

```python
text1 = 'Appple services and Products https://www.apple.com'
text2 = 'Wesite builders used to improve domain https://websitebuilders.com'
text3 = 'Welcome to Google notebook www.google.com/notebook'
text4 = "Tokenization and lemmatization www.geeksforgeeks.org/nlp-how-tokenizing-text-sentence-words-works/"
```

```python
remove_url(text4)
```
