```python
# Import library
from nltk.tokenize import wordpunct_tokenize
```

```python
# English to Hindi(Hinglish) Dictionary

translation_dict = {
    "i": "mai",
    "am": "hoon",
    "going": "jaa raha",
    "home": "ghar",
    "you": "tum",
    "are": "ho",
    "hello": "namaste",
    "my": "mera",
    "name": "naam",
    "is": "hai",
    "good": "accha",
    "boy": "ladka",
    "girl": "ladki",
    "food": "khana",
    "water": "paani",
    "come": "aao",
    "student": "vidyarthi"
}
```

```python
# Translation Function

def translate_to_hindi(sentence):

    # Tokenize sentence
    words = wordpunct_tokenize(sentence.lower())

    translated_words = []

    # Translate word by word
    for word in words:

        translated_word = translation_dict.get(word, word)

        translated_words.append(translated_word)

    # Join translated words
    return " ".join(translated_words)
```

```python
# User Input

sentence = input("Enter English Sentence: ")

translated = translate_to_hindi(sentence)

print("\nTranslated Sentence:")
print(translated)
```
