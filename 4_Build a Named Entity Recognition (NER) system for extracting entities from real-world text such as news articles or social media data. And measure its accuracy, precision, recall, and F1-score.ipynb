```python
# Simple NER System using spaCy
# Install:
# pip install spacy scikit-learn
# python -m spacy download en_core_web_sm

import spacy
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load English NER model
nlp = spacy.load("en_core_web_sm")

# Sample real-world text
texts = [
    "Elon Musk founded SpaceX in California.",
    "Virat Kohli plays cricket in India.",
    "Apple launched iPhone 15 in New York."
]

# Actual entities (Ground Truth)
# PERSON, ORG, GPE = Location
true_labels = [
    ["PERSON", "ORG", "GPE"],
    ["PERSON", "GPE"],
    ["ORG", "PRODUCT", "GPE"]
]

predicted_labels = []

# Perform NER
for text in texts:
    doc = nlp(text)

    entities = []
    print("\nText:", text)

    for ent in doc.ents:
        print(ent.text, "->", ent.label_)
        entities.append(ent.label_)

    predicted_labels.append(entities)

# Flatten lists for evaluation
y_true = []
y_pred = []

for t, p in zip(true_labels, predicted_labels):
    min_len = min(len(t), len(p))

    y_true.extend(t[:min_len])
    y_pred.extend(p[:min_len])

# Evaluation Metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='micro')
recall = recall_score(y_true, y_pred, average='micro')
f1 = f1_score(y_true, y_pred, average='micro')

# Results
print("\nEvaluation Metrics")
print("Accuracy :", round(accuracy, 2))
print("Precision:", round(precision, 2))
print("Recall   :", round(recall, 2))
print("F1-Score :", round(f1, 2))
```
