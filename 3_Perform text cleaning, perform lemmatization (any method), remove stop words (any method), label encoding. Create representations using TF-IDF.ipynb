```python
# Install once
# pip install nltk scikit-learn pandas

import pandas as pd
import nltk
import string

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer

# Download required files (Run once)
nltk.download('stopwords')
nltk.download('wordnet')

# ---------------------------------------------------
# Sample Data
# ---------------------------------------------------

data = {
    "Text": [
        "I love Machine Learning!",
        "Python is very easy to learn.",
        "NLP is interesting and useful."
    ],
    "Category": ["Positive", "Positive", "Neutral"]
}

df = pd.DataFrame(data)

# ---------------------------------------------------
# 1. Text Cleaning
# ---------------------------------------------------

def clean_text(text):
    text = text.lower()  # lowercase
    text = text.translate(str.maketrans('', '', string.punctuation))  # remove punctuation
    return text

df["Clean_Text"] = df["Text"].apply(clean_text)

# ---------------------------------------------------
# 2. Remove Stop Words
# ---------------------------------------------------

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    words = [word for word in words if word not in stop_words]
    return " ".join(words)

df["No_Stopwords"] = df["Clean_Text"].apply(remove_stopwords)

# ---------------------------------------------------
# 3. Lemmatization
# ---------------------------------------------------

lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    words = text.split()
    words = [lemmatizer.lemmatize(word) for word in words]
    return " ".join(words)

df["Lemmatized_Text"] = df["No_Stopwords"].apply(lemmatize_text)

# ---------------------------------------------------
# 4. Label Encoding
# ---------------------------------------------------

encoder = LabelEncoder()

df["Label"] = encoder.fit_transform(df["Category"])

# ---------------------------------------------------
# 5. TF-IDF Representation
# ---------------------------------------------------

tfidf = TfidfVectorizer()

tfidf_matrix = tfidf.fit_transform(df["Lemmatized_Text"])

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf.get_feature_names_out()
)

# ---------------------------------------------------
# Save Outputs
# ---------------------------------------------------

# Save cleaned data
df.to_csv("processed_text_data.csv", index=False)

# Save TF-IDF data
tfidf_df.to_csv("tfidf_output.csv", index=False)

# ---------------------------------------------------
# Display Results
# ---------------------------------------------------

print("Processed Data:")
print(df)

print("\nTF-IDF Representation:")
print(tfidf_df)

print("\nFiles Saved:")
print("1. processed_text_data.csv")
print("2. tfidf_output.csv")
```
